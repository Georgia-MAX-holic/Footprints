# Bias_ Variance Trad-off(편향-분산 트레이드 오프)

- Bias, Variance : 무언가를 학습시키고 예측할 때 생기는 세가지, 혹은 두가지 오차

- Trad-Off 관계: Bias가 올라가면 Variance가 내려가는 관계 



img1.daumcdn.png

- 타겟 한 가운데를 실제 값이라고 한다면, 점찍은 부분은 예측한 부분 .

-  Bias 오차가 높다는 것은 실측치와 예측치간의 오차가 크게 벌어진것 

- Variance 오차가 높다는 것은 예측의 범위가 넓다는 것을 말함 

즉 

- Bias 오차가 높다는 것은 과소적합한 상황 ( 꾸준히 틀림 ) 
- Variance 오차가 높다는 것은 과대적합한 상황 ( 예측 분포가 큰 상황



---

### 학습을 시킬수록 모델 복잡도는 올라감 

- 학습을 시킬수록 ( 모델 복잡도가 올라갈 수록) Bias(편향)은 줄어드나 , Variance(분산)은 올라가는것을 말함 

- 학습이 덜 될수록(모델 복잡도가 낮을수록) Bias(편향)은 올라가나 Variance(분산)은 내려가는것을 말함 



### 요약하자면...

- Bias 에러가 높아지는것은 많은 데이터를 고려하지 않아 ( 모델이 너무 단순해) 정확한 예측을 하지 못하는 경우를 말하고 

- Variance(분산) 에러는 노이즈까지 전부 학습해( 모델이 너무 복잡해) 약간의 input에도 예측 Y 값이 크게 흔들리는 것을 말함 

- 이 두가지 에러가 상호 Trade-off 관계에 있어서 이 둘을 모두 잡는 것은 불가능 한 딜레마가 생김 

- 결국 기대되는 MSE의 값은 Variance와 Bias간의 Trade-Off 때문에 계속 학습 시킨다고 해서 전체 Error가 줄어드는 것은 아님 



# **이를 위해서 Validation set 이라는 데이터 셋 필요**

- 모델을 한번 학습 시켰다면 그것을 Validation Set으로 에러율이 얼마나 내려가는지 한번 더 확인해 보는것 

- Validation Set이 최적의 Error율을 보인다면 그 지점이 가장 학습이 잘 된 지점으로 생각하고 학습을 중단 시키는것
