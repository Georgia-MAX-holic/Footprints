위와같이 신경망은 노드가 가중치로 연결되어 입력 신호와 연산한 뒤에 출력값으로 내보내는 함수라고도 할 수 있음 

- 학습 : 적절한 가중치를 찾는 과정 ( Training, learning) 

- 입력층 (Input Layer): 데이터가 입력되는 층, 입력되는 데이터셋의 특성(Feature)에 따라 입력층 노드의 수가 결정됨 보통 입력층은 어떤 계산도 수행하지 않고 그냥 값들을 전달하기만 하는 특징을 가지고 있음 .
그렇기 때문에 신경망의 층수(깊이 , Depth)를 셀 때 입력층은 포함하지 않음

- 은닉층 :
은닉층은 입력층으로부터 입력된 신호가 가중치, 편향과 연산되는 층 
일반적으로 입력층과 출력층 사이에 있는 층을 은닉층,
은닉층에서 일어나는 계산의 결과를 사용자가 볼 수 없기에 은닉층 이라는 이름

입력 데이터셋의 특성 수와 상관없이 노드 수를 구성할 수 있음 

일반적으로 딥러닝(Deep learning)이라고 하면 2개 이상의 은닉층을 가진 신경망 



- 출력층 :
가장 마지막에 위치한 층이며 은닉층 연산을 마친 값이 출력되는 층 
출력층을 잘 설계하는것이 중요

1. 이진 분류(Binary Classification) : 활성화 함수로는 시그모이드 함수를 사용하며 출력층의 노드는 1로 설정 
출력되는 갓이 0과 1사이의 확률값이 되도록 함 

2. 다중 분류 (Multi-class Classification) : 활성화 함수로는 소프트맥스(SoftMax) 함수를 사용하며 출력층의 노드 수는 레이블의 클래스 수와 동일하게 설정

3. 회귀(Regression : 일반적으로는 활성화 함수를 지정해주지 않으며 출력층의 노드 수는 출력값의 특성(Feature) 수와 동일하게 설정 
ex)} 단순히 하나의 수를 예측하는 문제라면 1
