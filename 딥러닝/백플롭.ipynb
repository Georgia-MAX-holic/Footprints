```python
import numpy as np

np.random.seed(812)
```

```python
# [.  ]
X = np.array((8,8],
              [2.5],
              [7,6]), dtype = float)

#        
y = X[:,0]*5 + X[:,1]*2
y = y.reshape(3,1)
y = y.reshape(3,1)

```

 (Normalization) 

```python
       0~1   

X = X/ np.amax(X , axis=0)
y = y/ np.amax(y, axis=0)

print(" ,  \m", X)
print(" \n", y)
```
