- 예제 
공부 시간과 수면시간을 특성으로 하고 시험 점수를 레이블로 하는 회귀 예제 

아래 식에서 x1은 공부시간을 나타내고 x2 는 수면시간을 나타냄 

$y = 5x_1 + 2x_2 + 40$

```python
import numpy as np

np.random.seed(812)
```

```python
# [공부시간. 수면시간 ]
X = np.array((8,8],
              [2.5],
              [7,6]), dtype = float)

# 선형 관계를 바탕으로 시험 점수 레이블 생성 
y = X[:,0]*5 + X[:,1]*2
y = y.reshape(3,1)
y = y.reshape(3,1)

```

특성 정규화(Normalization) 

```python
각 특성 및 레이블의 최댓값으로 나누어 주어 0~1사이의 값으로 만듬 

X = X/ np.amax(X , axis=0)
y = y/ np.amax(y, axis=0)

print(" 공부시간, 수면시간 \m", X)
print("시험점수 \n", y)
```
