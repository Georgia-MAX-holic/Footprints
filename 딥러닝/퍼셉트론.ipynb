# **SGD (Stochastic Gradient Descent) 와 Mini-Bat**ch GD

- Gradient Descent 방식은 전체 학습 데이터를 기반으로 GD를 계산. 하지만 입력 데이터가 크고 레이어가 많을 수록 GD를 계산하는데 많은 Computing 자원이 소모( 일반적으로 메모리 부족으로 연산 불가능할 수 있음)

- 이를 극복하기 위해 Stochastic Gradient Descent와 Mini-Batch Gradient Descent 방식이 도입 

 

```python
GD(Gradient Descent)        | SGD                               | Mini -Batch GD 
-----------------------------------------------------------------------
전체 학습 데이터를 기반으로 |전체 학습 데이터 중                |전체 학습 데이터 중 특정 크기 만큼(Batch 크기)
GD를 계산                   | 한 건만 임의로 선택하여 GD를 계산 |임의로 선택해서 GD 계산


[일반적으로 Mini-Batch GD가 대부분의 딥러닝 Framework에서 채택됨 ]
```

# **1. Batch Gradient Descent **
 - 전체 학습 데이터의 Gradient를 평균내어 GD의 매번 step마다 적용 

# **2. Stochastic Gradient Descent** 
 - 데이터 셋이 매우 큰경우, 매번 전체 학습 데이터를 연산하는 것은 메모리 사용량이 늘어나는 일 

 - 하나의 데이터를 골라 인공신경망에 학습시킴 

 - Full batch와는 다르게 확률론적, 불확실성의 개념을 도입 


optimiztion.의 loop 

 - Take an example
 - Feed it to Neural Netwrok
 - Calculate it's gradient
 - Use the gradient we calculated in step 3 to update the weights
 - Repeat steps 1-4 for all the examples in training dataset 



# **3. SGD 장점** 

  - SGD는 빠른 속도로 수렴 
  - SGD는 local optima에 빠질 위험이 적음 


# 4. mini-Batch Gradient Descent **굵은 텍스트**

 - 위 두 방법의 절충안 
 - 전체 학습 데이터를 minibatch로 나누어, GD 진행 

optimization의 loop 는 이와 같다

 - Pick a mini-batch (하나의 데이터가 아닌)
 - Feed it to Neural Network
 - Calculate the mean gradient of the mini-batch (batch GD의 특성 적용)

 - Use the mean gradient we caculated in step 3 to update the weights 

 -Repeat steps 1-4 for the mini-batches we created 


# **minibatch GD 장점** 

 - SGD 에 비해 local optima에 빠질 위험이 줄어듬 
 - 병렬처리에 유리(Batch를 나누니까)
 - Batch GD보다 메모리 사용을 절약 가능

# **1. 가중치(Weight) - 편향(bias) 연산**

    퍼셉트론은 여러 개의 신호를 입력받음
    입력된 신호는 각각의 가중치와 곱해지고, 그 결과를 더해주게 됨
    이러한 가중합(Weighted Sum)이 퍼셉트론의 첫 번째 단계임 

```python
import numpy as np

input = np.array([1, 2, 3])
weight = np.array([0.2, 0.3, -0.1])

np.dot(input, weight)
```

위 식에서 입력신호 (input) [1,2,3] 에 대해 임의의 가중치(Weight) [0.2, 0.3, -0.1]가 설정되있고 계산식은 

- 1×0.2+2×0.3+3×(−0.1)=0.5

- ∑(b+w0x0+w1x1+...+wnxn)

- 시그모이드는 이러한 문제점을 해결하기 위하여 탄생

# **인공 신경망** 

- 딥러닝 : 인공 신경망의 층을 깊게 쌓은 것

- ANN(Artificail Neural Networks), 즉 인공 신경망은 실제 신경계를 모사하여 만들어진 계산 모델 , 줄여서 뉴럴넷 

- 여러개의 층으로 쌓아 구축한 신경망 :다층 퍼셉트론 신경망 ( Multi-Layer Perceptron, MLP)



ADcl9EN.png

# - **입력층 (Input Layer) :**
   1. 데이터셋이 입력되는 층, 입력되는 데이터셋의 특성(Feature)에 따라 입력층의 노드의 수가 결정됨.

   2. 보통 입력층은 어떤 계산도 수행하지 않고 그냥 값들을 전달하기만 하는 특징

# - **은닉층(Hidden Layers):**
   1. 입력층으로부터 입력된 신호가 가중치, 편향과 연산되는 층.

   2. 일반적으로 입력층과 출력층 사이에 있는 층을 은닉층 이라고 함 

   3. 은닉층의 계산 결과를 사용자가 볼 수 없기 때문에 [은닉]

   4. 일반적인 딥러닝 : 2개 이상의 은닉층을 가진 신경망 


# - **출력층 (Output Layer):**


    가장 마지막에 위치한 층, 은닉층 연산을 마친 값이 출력되는 층 , 해결해야 할 문제에 따라서 출력층을 잘 설계하는것이 중요 

   1. 이진 분류(Binary Classification) : 활성화 함수로는 시그모이드(Sigmoid) 함수를 사용하며 출력층의 노드 수는 1로 설정 , 출력되는 값이 0과 1 사이의 확률값이 되도록 함 

   2. 다중 분류 (Multi-class Classification): 활성화 함수로는 소프트맥스(Softmax)함수를 사용하며 출력층의 노드 수는 레이블의 클래스 수와 동일하게 설정함 

   3. 회귀(Regression) : 일반적으로는 활성화 함수를 지정해주지 않으며 출력층의 노드 수는 출력값의 특성(Feature)수와 동일하게 설정함 

   단순히 하나의 수를 예측하는 문제면 1로 해주면 됨

```python
# Iris 분류 
# 전체 특성 중 (sepal length, petal length)만 사용 
#150개 데이터중 Setosa(50), Versicolor(50)만 추출하여 100개의 데이터에 대하여 이진 분류 (Binary classfication) 실행

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
label = df.iloc[0:100 , 4].values
label
```

```python
label = np.where(label == 'Iris-setosa' , 0,1)  #np.where 인덱스만 반환  #  Iris-setosa 면 0 , 아니면 1 
label.astype(object)
label
```

```python
features = df.iloc[0:100, [0,2]].values
features.shape
```

```python
plt.scatter(features[:50,0],features[:50,1], color='red' , marker='o', label='setosa')
plt.scatter(features[50:100, 0], features[50:100, 1], color='blue' ,marker='x', label ='versicolor')
plt.xlabel("sepal length")
plt.ylabel("petal lenght")
plt.legend(loc='upper left')
plt.show()
```

```python
# 학습 데이터셋, 시험 데이터셋으로 나눔 

from sklearn.model_selection import train_test_split

X_train, X_test , y_train , y_test = train_test_split(features, label , test_size = 0.2 , random_state =42)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, activation ='sigmoid')
])
```

```python
model.compile(optimizer='sgd',
              loss="binary_crossentropy",
              metrics=['accuracy'])
```

```python


model.fit(X_train, y_train, epochs=30)

```

```python
model.evaluate(X_test, y_test , verbose=2)
```
